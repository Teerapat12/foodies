## Foodies Shop Ranker

---

The code logic is hidden in foodies/shop_ranker.py (ShopRanker) class

Here the class `ShopRanker` would accept shop dataframe, weight of scoring,and distance_scorer

Then we can call `get_top_rank` function to get the top N shop which obtained the highest score based on our logic.

```python
from foodies.shop_ranker import ShopRanker
import pandas as pd
```

```python
shop = pd.read_csv("./data/preped_shop").drop(['Unnamed: 0','scores_breakdown','scores'], axis = 1)

first_weight =  {
     'tel_count': 0.3,
     'has_website': 0.8,
     'has_line': 0.6,
     'has_facebook':0.4,
     'has_instagram': 0.7,
     'has_all': 0.3,
     'has_carpark_i':0.9,
     'open_period_count': 0.2,
     'shop_average_log_price': -0.75,
     'near_shop_counts_log': 0.03,
     'description_length': 0.05
}

def distance_score_calculator(score):
    if(score<1):
        return 1.1
    elif(score<5):
        return 1
    else:
        return 0.8
```

## Try with given distance scorer

```python
shop_ranker = ShopRanker(shop, first_weight, distance_scorer=distance_score_calculator)
shop_ranker.get_top_rank(10, {"lat":13.755376,"lng":100.529406})
```

```python
shop_ranker.get_top_rank(10, {"lat":172.819182,"lng":90.283171})
```
